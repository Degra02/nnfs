```python
from Layer_Dense import LayerDense
from Activation_ReLU import ActivationReLU
from CreateData import create_data
import numpy as np
import matplotlib.pyplot as plt
plt.rcParams['figure.figsize'] = [10, 10]
np.random.seed(0)
```

Various imports, then creating a random input (input layer)

```python
X, y = create_data(100, 3)
plt.scatter(X[:, 0], X[:, 1], c=y, cmap="brg")
plt.show()
```

Initializing the first layer, then apllying the ReLU function

```python
layer1 = LayerDense(2, 5)
activation1 = ActivationReLU()

layer1.forward(X)
activation1.forward(layer1.output)
activation1.output
```

```python
plt.scatter(activation1.output[:, 0], activation1.output[:, 1], c=y, cmap="brg")
plt.show()
```
